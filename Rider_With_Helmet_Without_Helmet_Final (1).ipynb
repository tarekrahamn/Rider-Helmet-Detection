VGG16 Model

```python
from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession

config = ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.5
config.gpu_options.allow_growth = True
session = InteractiveSession(config=config)
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
# import the libraries as shown below

from tensorflow.keras.layers import Input, Lambda, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg19 import VGG19
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
import numpy as np
from glob import glob
#import matplotlib.pyplot as plt
```

```python
# re-size all the images to this
IMAGE_SIZE = [640, 640]

train_path = '/content/YOLOv8_Segmentation_DeepSORT_Object_Tracking/ultralytics/yolo/v8/segment/Segmentation_DIP-6/train'
valid_path = '/content/YOLOv8_Segmentation_DeepSORT_Object_Tracking/ultralytics/yolo/v8/segment/Segmentation_DIP-6/valid'
```

```python
# Import the VGG16 library as shown below and add preprocessing layer to the front of VGG
# Here we will be using imagenet weights

vgg16 = VGG16(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
# don't train existing weights
for layer in vgg16.layers:
    layer.trainable = False
```

```python
  # useful for getting number of output classes
folders = glob('/content/YOLOv8_Segmentation_DeepSORT_Object_Tracking/ultralytics/yolo/v8/segment/Segmentation_DIP-6/train/*')
```

```python
folders
```

```python
# our layers - you can add more if you want
x = Flatten()(vgg16.output)
```

```python
len(folders)
```

```python
prediction = Dense(len(folders), activation='softmax')(x)

# create a model object
model = Model(inputs=vgg16.input, outputs=prediction)
```

```python
# view the structure of the model
model.summary()
```

```python
# tell the model what cost and optimization method to use
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

```python
# Use the Image Data Generator to import the images from the dataset
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
# Make sure you provide the same target size as initialied for the image size
training_set = train_datagen.flow_from_directory('/content/YOLOv8_Segmentation_DeepSORT_Object_Tracking/ultralytics/yolo/v8/segment/Segmentation_DIP-6/train',
                                                 target_size = (640, 640),
                                                 batch_size = 8,
                                                 class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory('/content/YOLOv8_Segmentation_DeepSORT_Object_Tracking/ultralytics/yolo/v8/segment/Segmentation_DIP-6/test',
                                            target_size = (640, 640),
                                            batch_size = 8,
                                            class_mode = 'categorical')
```

Train with VGG16

```python
from keras.layers import Dense
from keras.models import Model


input_layer = model.input


last_hidden_layer = model.layers[-2].output

# Create a new output layer (change the number of units to the number of classes)
output_layer = Dense(2, activation='softmax')(last_hidden_layer)

# Create a new model with the same input but new output layer
new_model = Model(inputs=input_layer, outputs=output_layer)

# Now compile and fit the new model as usual
new_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
r = new_model.fit(
    training_set,
    validation_data=test_set,
    epochs=2,
    steps_per_epoch=len(training_set),
    validation_steps=len(test_set)
)
```

Face Extractions Test

```python
import cv2
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Load and preprocess the image
img_path = "/content/YOLOv8_Segmentation_DeepSORT_Object_Tracking/ultralytics/yolo/v8/segment/Segmentation_DIP-6/test/images/14316725_1753554428228371_3725241226161575478_n_jpg.rf.a80ac34abd769e63d4cae7b49070f202.jpg"
img_pred = image.load_img(img_path, target_size=(640, 640))
img_array = image.img_to_array(img_pred)
img_array = np.expand_dims(img_array, axis=0)

# Predict the class probabilities
predictions = model.predict(img_array)

# Get the classes and their corresponding probabilities
class_names = ["Rider With Helmet", "Rider Without Helmet"]  # Update with your class names
pred_classes = []
for i, pred in enumerate(predictions[0]):
    if pred > 0.7:
        pred_classes.append(class_names[i])

# Display the image with the predicted classes as titles
plt.imshow(img_pred)
plt.title(f"Predictions: {', '.join(pred_classes)}")
plt.axis('off')
plt.show()

# If "Rider Without Helmet" is predicted, extract and display the face of the rider
if "Rider Without Helmet" in pred_classes:
    # Load the pre-trained face detection model
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # Convert the image to grayscale
    gray_img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2GRAY)

    # Detect faces in the grayscale image
    faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Extract and display the face regions without upscaling
    margin = 20  # Add a margin around the detected face region
    for (x, y, w, h) in faces:
        # Expand the detected face region with a margin
        x_margin = max(0, x - margin)
        y_margin = max(0, y - margin)
        w_margin = min(img_array.shape[2], w + 2 * margin)
        h_margin = min(img_array.shape[1], h + 2 * margin)

        # Extract the expanded face region
        face_img = img_array[0, y_margin:y_margin+h_margin, x_margin:x_margin+w_margin, :]

        # Display the face of the rider without upscaling
        plt.imshow(face_img.astype(np.uint8))
        plt.title("Face of Rider Without Helmet")
        plt.axis('off')
        plt.show()
```
